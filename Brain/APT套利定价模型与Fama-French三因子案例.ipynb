```python
#载入需要用到的函数库
import numpy as np 
import pandas as pd 
#筛选排序SL,SM,SH,BL,BM,BH
marketstock=pd.read_excel('/Users//stanley//Desktop//上证50.xlsx','市值与PB',header=0,index_col=0)
marketstock=marketstock.dropna()  #删除缺省的数据
marketstock
```

```python

S=marketstock.sort_values(["marketvalue"],ascending=True).head(25) #获取市值《50% 正数25支
B=marketstock.sort_values(["marketvalue"],ascending=True).tail(24) #获取市值>50%  倒数24支
BP30=np.percentile(marketstock['1/PB'],30) 30%分位
BP70=np.percentile(marketstock['1/PB'],70) 70%分位
SL_value=S.loc[(S["1/PB"]<=BP30)] 
SM_value=S.loc[(S["1/PB"]>BP30)& (S["1/PB"]<=BP70)]    
SH_value=S.loc[(S["1/PB"]>BP70)]    
BL_value=B.loc[(B["1/PB"]<=BP30)]
BM_value=B.loc[(B["1/PB"]>BP30)& (B["1/PB"]<=BP70)]    
BH_value=B.loc[(B["1/PB"]>BP70)]  

```

```python
SL_value
```

```python

SL_weight=SL_value.marketvalue/sum(SL_value.marketvalue) #求取SL各股票加权平均值
SL=pd.read_excel('/Users//stanley//Desktop//上证50.xlsx','SL',header=0,index_col=0)
```

```python
SL
```

```python

SL_log=np.log(SL/SL.shift(1))  #按照对数收益率公式计算对数收益率
SL_log=SL_log.dropna()  #删除缺省的数据
SL_weight_log=SL_log*SL_weight
r_SL= SL_weight_log.apply(lambda x: x.sum(), axis=1)  #SL组加权收益率
```

```python
SM_weight=SM_value.marketvalue/sum(SM_value.marketvalue) #求取SM各股票加权平均值
SM=pd.read_excel('/Users//stanley//Desktop//上证50.xlsx','SM',header=0,index_col=0)
SM_log=np.log(SM/SM.shift(1))  #按照对数收益率公式计算对数收益率
SM_log=SM_log.dropna()  #删除缺省的数据
SM_weight_log=SM_log*SM_weight
r_SM= SM_weight_log.apply(lambda x: x.sum(), axis=1)  #SM组加权收益率

SH_weight=SH_value.marketvalue/sum(SH_value.marketvalue) #求取SH各股票加权平均值
SH=pd.read_excel('/Users//stanley//Desktop//上证50.xlsx','SH',header=0,index_col=0)
SH_log=np.log(SH/SH.shift(1))  #按照对数收益率公式计算对数收益率
SH_log=SH_log.dropna()  #删除缺省的数据
SH_weight_log=SH_log*SH_weight
r_SH= SH_weight_log.apply(lambda x: x.sum(), axis=1)  #SH组加权收益率

BL_weight=BL_value.marketvalue/sum(BL_value.marketvalue) #求取BL各股票加权平均值
BL=pd.read_excel('/Users//stanley//Desktop//上证50.xlsx','BL',header=0,index_col=0)
BL_log=np.log(BL/BL.shift(1))  #按照对数收益率公式计算对数收益率
BL_log=BL_log.dropna()  #删除缺省的数据
BL_weight_log=BL_log*BL_weight
r_BL=BL_weight_log.apply(lambda x: x.sum(), axis=1)  #BL组加权收益率

BM_weight=BM_value.marketvalue/sum(BM_value.marketvalue) #求取BM各股票加权平均值
BM=pd.read_excel('/Users//stanley//Desktop//上证50.xlsx','BM',header=0,index_col=0)
BM_log=np.log(BM/BM.shift(1))  #按照对数收益率公式计算对数收益率
BM_log=BM_log.dropna()  #删除缺省的数据
BM_weight_log=BM_log*BM_weight
r_BM=BM_weight_log.apply(lambda x: x.sum(), axis=1)  #BM组加权收益率

BH_weight=BH_value.marketvalue/sum(BH_value.marketvalue) #求取BH各股票加权平均值
BH=pd.read_excel('/Users//stanley//Desktop//上证50.xlsx','BH',header=0,index_col=0)
BH_log=np.log(BH/BH.shift(1))  #按照对数收益率公式计算对数收益率
BH_log=BH_log.dropna()  #删除缺省的数据
BH_weight_log=BH_log*BH_weight
r_BH=BH_weight_log.apply(lambda x: x.sum(), axis=1)  #BH组加权收益率

SMBr=(r_SL+r_SM+r_SH)/3-(r_BL+r_BM+r_BH)/3    #当年的SMB和HML因子，存为list
HMLr=(r_SH+r_BH)/2-(r_SL+r_BL)/2  
```

```python

sz50_index=pd.read_excel('/Users//stanley//Desktop//上证50.xlsx','上证50指数收盘价',header=0,index_col=0)
sz50_index_log=np.log(sz50_index/sz50_index.shift(1))  #按照对数收益率公式计算对数收益率
sz50_index_log=sz50_index_log.dropna()  #删除缺省的数据
ret_merge=pd.merge(pd.DataFrame(sz50_index_log),pd.DataFrame(SMBr),left_index=True,right_index=True,how='inner') 
ret_merge=pd.merge(pd.DataFrame(ret_merge),pd.DataFrame(HMLr),left_index=True,right_index=True,how='inner') 
ret_merge.columns=['sz50_index_log','SMBr','HMLr']
ret_merge
```

```python
ret_merge.corr()  #三个变量的相关系数
```

```python

#计算日无风险利率  
Rf_year=0.03 #以中国三年期国债年化收益率为无风险利率  
Rf=(1+Rf_year)**(1/365)-1 ##年利率转化为日利率 每日复利 
#计算风险溢价:Ri-Rf  
Eret=ret_merge-Rf  
Eret.head() 
```

```python

##利用最小二乘法进行线性回归，拟合FF模型 （以三一重工、工商银行为例）
import statsmodels.api as sm  
model = sm.OLS(SL_log[['三一重工']], sm.add_constant(Eret[['sz50_index_log', 'SMBr', 'HMLr']].values))
result = model.fit()
result.summary()  
print(result.summary())
```

```python
【提示】这里x1、x2、x3分别对应了市场风险溢酬因子、市值因子、账面市值比因子，const对应着常数项，即超额收益率α。一般情况下，
当p值小于0.05时，我们认为该因子对于收益率有显著影响。R²是指拟合优度,是回归直线对观测值的拟合程度，R²越大说明拟合效果越好。
```
